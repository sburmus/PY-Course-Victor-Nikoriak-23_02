# 📦 Фінальний проєкт — Таємнича коробка Нової Пошти

Кожна група вирішує своє завдання і **лідер групи** відправляє результат.

```
         📦 BOX
           ↑
  ┌────────┼────────┬────────┐
  Група 1  Група 2  Група 3  Група 4
  Адреси   Вага    Логістика Відправка
```

> ⚠️ Якщо хоча б одна група помилиться — коробка не відкриється.

````python
# @title 🔒 Натисніть RUN
# 🔒 system cell (не змінювати)
# 🔒 SYSTEM CELL — запустити один раз

import requests
import csv
import io
import time

SYSTEM_READY = True
STUDENT_NAME = None

FINAL_FORM_URL = "https://docs.google.com/forms/d/e/1FAIpQLSe5oUb8rI8lz5ud_oCQDvj0_s6FUYHjFU-XitEgb94z0s3zdQ/formResponse"
FINAL_SHEET_CSV = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQ16oP8S-HwlXJsWhXtXpeUdIdZHV1mBnn9ipYQe0Dz57W4BnvGxkBG11NRatXY49SVOiGKFnFYe0Aw/pub?gid=2126041300&single=true&output=csv"

def require_student():
    if STUDENT_NAME is None or str(STUDENT_NAME).strip() == "":
        print("⛔ Спочатку представтесь у клітинці 👤")
        return False
    return True

def submit_group_result(group, value):
    """Відправляє результат групи у Google Sheet."""
    if not require_student():
        return

    data = {
        "entry.575335391": STUDENT_NAME,
        "entry.1320852544": "FinalProject",
        "entry.705949827": "group_result",
        "entry.678121911": group,
        "entry.856775041": str(value)
    }

    try:
        requests.post(FINAL_FORM_URL, data=data)
        print(f"✅ Результат {group} відправлено: {value}")
    except Exception as e:
        print(f"❌ Помилка відправки: {e}")

def read_group_results():
    """Читає останні результати кожної групи зі Sheet."""
    try:
        response = requests.get(FINAL_SHEET_CSV, timeout=10)
        reader = csv.DictReader(io.StringIO(response.text))
        results = {}
        for row in reader:
            group = row.get("group", "").strip()
            result = row.get("result", "").strip()
            if group.startswith("GROUP"):
                results[group] = result  # перезаписує → залишається останній
        return results
    except Exception as e:
        print(f"❌ Помилка читання Sheet: {e}")
        return {}

print("🔧 System ready ✅")
````
