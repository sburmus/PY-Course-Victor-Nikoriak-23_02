Клас Особа

Створіть клас під назвою Person. Зробіть так, щоб метод __init__() приймав ім'я, прізвище та вік як параметри та додавав їх як атрибути. Створіть ще один метод під назвою talk(), який виводить привітання від особи, наприклад, такого вигляду: "Привіт, мене звати Карл Джонсон, і мені 26 років".



```python
class Person:
    def __init__(self,name,surname,age):
        self.name = name
        self.surname = surname
        self.age = age
    def talk(self):
        print(f"Hello, I'm {self.name} {self.surname} and I'm {self.age} years old.")
person1 = Person("Karl","Jondon",26)
person2 = Person("Anna","Smith",30)
person1.talk()
person2.talk()
```

Створіть клас Dog з атрибутом класу 'age_factor', що дорівнює 7. Створіть __init__(), який приймає значення віку собаки. Потім створіть метод `human_age`, який повертає вік собаки в людському еквіваленті.



```python
class Dog:
    age_factor = 7
    def __init__(self,age):
         self.age = age
    def human_age(self):      
         return self.age * self.age_factor
dog1 = Dog(3)
dog2 = Dog(5)
print(f"Dog1 is {dog1.human_age()} years old in human years.")
print(f"Dog2 is {dog2.human_age()} years old in human years.")
```
