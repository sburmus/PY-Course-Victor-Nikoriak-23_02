Створити клас Counter, підклас threading.Thread, з двома глобальними змінними: counter (початкове значення 0) та rounds (100 000). Реалізувати метод run(), який у циклі for 100 000 разів збільшує counter на 1. Створити 2 екземпляри потоку, запустити їх, об'єднати (join) та перевірити підсумкове значення лічильника — чи дорівнює воно 200 000? Пояснити отриманий результат.

```python
import threading

counter = 0
rounds = 100_000


class Counter(threading.Thread):
    def run(self):
        global counter
        for _ in range(rounds):
            counter += 1


if __name__ == "__main__":
    t1 = Counter()
    t2 = Counter()

    t1.start()
    t2.start()

    t1.join()
    t2.join()

    print(f"Очікуваний результат: {2 * rounds}")
    print(f"Фактичний результат:  {counter}")
```

Створіть сервер socket echo, який обробляє кожне з'єднання в окремому потоку

```python
import socket
import threading

HOST = "127.0.0.1"
PORT = 8080

def handle_client(conn, addr):
    print(f"[+] Підключено: {addr}")
    try:
        while True:
            data = conn.recv(1024)
            if not data:
                break
            print(f"[{addr}] отримано: {data.decode()}")
            conn.sendall(data)
    finally:
        conn.close()
        print(f"[-] З’єднання закрито: {addr}")

def main():
    server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server.bind((HOST, PORT))
    server.listen()
    server.settimeout(1.0)  # таймаут на accept(), щоб не блокуватись назавжди
    print(f"Echo‑сервер запущено на {HOST}:{PORT}")

    try:
        while True:
            try:
                conn, addr = server.accept()
                thread = threading.Thread(target=handle_client, args=(conn, addr))
                thread.start()
            except socket.timeout:
                # кожну секунду перевіряємо, чи не було Ctrl+C
                continue
    except KeyboardInterrupt:
        print("\n[!] Сервер зупинено вручну")
    finally:
        server.close()

if __name__ == "__main__":
    main()
```
