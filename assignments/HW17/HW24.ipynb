Завдання 1

Напишіть програму, яка зчитує послідовність символів та друкує їх у зворотному порядку, використовуючи вашу реалізацію Stack.



```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.items:
            raise IndexError("Попитка pop з порожнього стеку")
        return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0


def reverse_string(s: str) -> str:
    stack = Stack()
    for ch in s:
        stack.push(ch)

    reversed_s = ""
    while not stack.is_empty():
        reversed_s += stack.pop()
    return reversed_s


# приклад
print(reverse_string("Світлана"))  
```

Завдання 2

Напишіть програму, яка зчитує послідовність символів та визначає, чи є її дужки, фігурні дужки та фігурні дужки «збалансованими».

```python
def is_balanced(s: str) -> bool:
    stack = []
    pairs = {')': '(', ']': '[', '}': '{'}

    for ch in s:
        if ch in "([{":
            stack.append(ch)
        elif ch in ")]}":
            if not stack or stack[-1] != pairs[ch]:
                return False
            stack.pop()
    return not stack


# приклади
print(is_balanced("()[]{}"))      
print(is_balanced("([{}])"))      
print(is_balanced("([)]"))       
print(is_balanced("((("))         
```

Розширте стек, включивши метод під назвою get_from_stack, який шукає та повертає елемент e зі стеку. Будь-який інший елемент повинен залишатися в стеку, дотримуючись їхнього порядку. Розглянемо випадок, коли елемент не знайдено - викличіть ValueError з належною інформацією Message.

 

Розширте Чергу, включивши метод під назвою get_from_stack, який шукає та повертає елемент e з черги. Будь-який інший елемент повинен залишатися в черзі, дотримуючись їхнього порядку. Розглянемо випадок, коли елемент не знайдено - викличіть ValueError з належною інформацією Message.

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.items:
            raise IndexError("Попитка pop з порожнього стеку")
        return self.items.pop()

    def is_empty(self):
        return len(self.items) == 0

    def get_from_stack(self, e):
        temp = []
        found = None
        while not self.is_empty():
            item = self.pop()
            if item == e and found is None:
                found = item
                break
            temp.append(item)

        # повертаємо назад інші елементи
        while temp:
            self.push(temp.pop())

        if found is None:
            raise ValueError(f"Елемент {e} не знайдено у стеку")
        return found
```
