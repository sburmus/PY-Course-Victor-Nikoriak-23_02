Розширити несортований список

Реалізуйте методи append, index, pop, insert для UnsortedList. Також реалізуйте метод slice, який прийматиме два параметри 'start' та 'stop' та повертатиме копію списку, починаючи з позиції та до позиції stop, але не включаючи її.

```python
class UnsortedList:
    def __init__(self):
        self.items = []

    def append(self, item):
        self.items.append(item)

    def index(self, item):
        if item in self.items:
            return self.items.index(item)
        raise ValueError(f"{item} не знайдено у списку")

    def pop(self, pos=None):
        if pos is None:
            return self.items.pop()
        return self.items.pop(pos)

    def insert(self, pos, item):
        self.items.insert(pos, item)

    def slice(self, start, stop):
        return self.items[start:stop]

    def __str__(self):
        return str(self.items)


# приклад
ul = UnsortedList()
ul.append(10)
ul.append(20)
ul.insert(1, 15)
print(ul)                
print(ul.slice(0, 2))    
print(ul.index(20))      
print(ul.pop())          
```

Реалізуйте стек за допомогою однозв'язного списку.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None

    def push(self, data):
        new_node = Node(data)
        new_node.next = self.top
        self.top = new_node

    def pop(self):
        if self.top is None:
            raise IndexError("Попитка pop з порожнього стеку")
        data = self.top.data
        self.top = self.top.next
        return data

    def peek(self):
        if self.top is None:
            return None
        return self.top.data

    def is_empty(self):
        return self.top is None


# приклад
stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)
print(stack.pop())  
print(stack.peek()) 
```

Реалізуйте чергу, використовуючи однозв'язний список.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:
            self.front = self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node

    def dequeue(self):
        if self.front is None:
            raise IndexError("Попитка dequeue з порожньої черги")
        data = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None
        return data

    def peek(self):
        if self.front is None:
            return None
        return self.front.data

    def is_empty(self):
        return self.front is None


# приклад
queue = Queue()
queue.enqueue("A")
queue.enqueue("B")
queue.enqueue("C")
print(queue.dequeue()) 
print(queue.peek())    
```
