Створіть метод класу з назвою `validate`, який має викликатися з методу `__init__` для перевірки параметра email, переданого конструктору. Логіка всередині методу `validate` може полягати в перевірці, чи є переданий параметр email коректним рядком електронної пошти.

Підтвердження електронної пошти:

Дійсний формат адреси електронної пошти 

Адреса електронної пошти 

```python
import re

class User:
    def __init__(self, email: str):
        # Виклик методу validate при створенні об'єкта
        self.email = self.validate(email)

    def validate(self, email: str) -> str:
        # Простий варіант перевірки через регулярний вираз
        pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
        if not re.match(pattern, email):
            raise ValueError("Некоректна адреса електронної пошти")
        return email

    def __str__(self):
        return f"Користувач з email: {self.email}"


# ✅ Приклад використання
try:
    user1 = User("svitlana@example.com")
    print(user1)  # Користувач з email: svitlana@example.com

    user2 = User("wrong-email")  # викличе ValueError
except ValueError as e:
    print("Помилка:", e)
```

Реалізуйте 2 класи, перший - Бос, а другий - Працівник.

Працівник має властивість 'boss', а її значення має бути екземпляром Boss.

Ви можете перепризначити це значення, але вам слід перевірити, чи нове значення є Boss. Кожен Boss має список своїх власних працівників. Вам слід реалізувати метод, який дозволяє додавати працівників до Boss. Вам не дозволено додавати екземпляри класу Boss до списку працівників безпосередньо через доступ до атрибута, натомість використовуйте геттери та сеттери!

Ви можете рефакторингувати існуючий код.

'''

id_ - це просто випадкове унікальне ціле число

 

Бос класу:

    def __init__(self, id_: int, name: str, company: str):

        self.id = id_

        self.name = ім'я

        self.company = компанія

        self.workers = []

 

Працівник класу:

    def __init__(self, id_: int, name: str, company: str, boss: Бос):

        self.id = id_

        self.name = ім'я

        self.company = компанія

        self.boss = бос

'''

```python
class Boss:
    def __init__(self, id_: int, name: str, company: str):
        self.id = id_
        self.name = name
        self.company = company
        self._workers = []  # приватний список

    def add_worker(self, worker):
        if isinstance(worker, Worker):
            self._workers.append(worker)
        else:
            raise ValueError("Можна додавати тільки екземпляри Worker")

    @property
    def workers(self):
        return self._workers

    def __str__(self):
        return f"Boss {self.name} ({self.company})"


class Worker:
    def __init__(self, id_: int, name: str, company: str, boss: Boss):
        self.id = id_
        self.name = name
        self.company = company
        self._boss = None
        self.boss = boss  # виклик сеттера

    @property
    def boss(self):
        return self._boss

    @boss.setter
    def boss(self, new_boss):
        if not isinstance(new_boss, Boss):
            raise ValueError("boss має бути екземпляром класу Boss")
        self._boss = new_boss
        new_boss.add_worker(self)

    def __str__(self):
        return f"Worker {self.name} ({self.company}), Boss: {self.boss.name}"
# Створюємо босів
boss1 = Boss(1, "Іван", "IT-Company")
boss2 = Boss(2, "Олена", "Design-Studio")

# Створюємо працівників
worker1 = Worker(101, "Світлана", "IT-Company", boss1)
worker2 = Worker(102, "Андрій", "IT-Company", boss1)

print(worker1)  # Worker Світлана (IT-Company), Boss: Іван
print(worker2)  # Worker Андрій (IT-Company), Boss: Іван

# Перепризначаємо босів
worker1.boss = boss2
print(worker1)  # Worker Світлана (IT-Company), Boss: Олена

# Перевіряємо список працівників у босів
print([w.name for w in boss1.workers])  # ['Андрій']
print([w.name for w in boss2.workers])  # ['Світлана']
```

Напишіть клас TypeDecorators, який має кілька методів для перетворення результатів функцій до заданого типу (якщо це можливо):

методи:

to_int

to_str

to_bool

до_числа

 

Не забудьте використовувати @wraps

'''

клас TypeDecorators:

    пропуск

 

@TypeDecorators.to_int

def нічого не робити(рядок: рядок):

    повернутий рядок

 

@TypeDecorators.to_bool

def do_something(рядок: рядок):

    повернутий рядок

 

стверджувати, що нічого не робиться('25') == 25

стверджувати, що do_something('True') є True

'''


```python
from functools import wraps

class TypeDecorators:
    @staticmethod
    def to_int(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            result = func(*args, **kwargs)
            try:
                return int(result)
            except ValueError:
                raise ValueError(f"Не вдалося перетворити '{result}' на int")
        return wrapper

    @staticmethod
    def to_str(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            result = func(*args, **kwargs)
            return str(result)
        return wrapper

    @staticmethod
    def to_bool(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            result = func(*args, **kwargs)
            if str(result).lower() in ["true", "1", "yes"]:
                return True
            elif str(result).lower() in ["false", "0", "no"]:
                return False
            else:
                raise ValueError(f"Не вдалося перетворити '{result}' на bool")
        return wrapper

    @staticmethod
    def to_float(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            result = func(*args, **kwargs)
            try:
                return float(result)
            except ValueError:
                raise ValueError(f"Не вдалося перетворити '{result}' на float")
        return wrapper


# ✅ Приклад використання
@TypeDecorators.to_int
def do_nothing(string: str):
    return string

@TypeDecorators.to_bool
def do_something(string: str):
    return string

assert do_nothing("25") == 25
assert do_something("True") is True
```
