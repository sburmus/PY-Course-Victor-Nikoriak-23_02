Виберіть своє рішення для однієї з вправ у цьому модулі. Розробіть тести для цього рішення та напишіть тести, використовуючи бібліотеку unittest. 

```python
def with_index(iterable, start=0):
    index = start
    for item in iterable:
        yield index, item
        index += 1
```

```python
#test
import unittest

class TestWithIndex(unittest.TestCase):
    def test_basic(self):
        self.assertEqual(list(with_index(["a", "b"])), [(0, "a"), (1, "b")])

    def test_start(self):
        self.assertEqual(list(with_index(["x", "y"], start=5)), [(5, "x"), (6, "y")])

    def test_empty(self):
        self.assertEqual(list(with_index([])), [])
```

Напишіть тести для програми «Телефонна книга», яку ви реалізували в модулі 1. Розробіть тести для цього рішення та напишіть тести, використовуючи бібліотеку unittest.

```python
class PhoneBook:
    def __init__(self):
        self.records = []

    def add_record(self, name, phone):
        self.records.append({"name": name, "phone": phone})

    def search_by_name(self, name):
        return [r for r in self.records if r["name"] == name]

    def delete_record(self, phone):
        self.records = [r for r in self.records if r["phone"] != phone]
```

```python
import unittest

class TestPhoneBook(unittest.TestCase):
    def setUp(self):
        self.pb = PhoneBook()

    def test_add_and_search(self):
        self.pb.add_record("Світлана", "12345")
        result = self.pb.search_by_name("Світлана")
        self.assertEqual(result[0]["phone"], "12345")

    def test_delete(self):
        self.pb.add_record("Андрій", "67890")
        self.pb.delete_record("67890")
        result = self.pb.search_by_name("Андрій")
        self.assertEqual(result, [])
```
