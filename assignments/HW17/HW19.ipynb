Створіть власну реалізацію вбудованої функції enumerate з назвою 'with_index', яка приймає два параметри: 'iterable' та 'start', значення за замовчуванням дорівнює 0. Поради: див. документацію до функції enumerate

```python
def with_index(iterable, start=0):
    index = start
    for item in iterable:
        yield index, item
        index += 1

# ✅ Приклад використання
for i, val in with_index(["a", "b", "c"], start=1):
    print(i, val)

```

Створіть власну реалізацію вбудованої функції range з назвою in_range(), яка приймає три параметри: «start», «end» та необов'язковий крок. Поради: Дивіться документацію до функції «range».

```python
def in_range(start, end, step=1):
    current = start
    while current < end:
        yield current
        current += step

# ✅ Приклад використання
for num in in_range(2, 10, 2):
    print(num)

```

Створіть власну реалізацію ітерованого об'єкта, який можна використовувати всередині циклу for-in. Також додайте логіку для отримання елементів за допомогою синтаксису квадратних дужок.

```python

class MyIterable:
    def __init__(self, data):
        self.data = data
        self.index = 0

    def __iter__(self):
        self.index = 0
        return self

    def __next__(self):
        if self.index >= len(self.data):
            raise StopIteration
        value = self.data[self.index]
        self.index += 1
        return value

    def __getitem__(self, key):
        return self.data[key]


# ✅ Приклад використання
items = MyIterable(["apple", "banana", "cherry"])

# Використання в циклі for-in
for item in items:
    print(item)

# Доступ через індекс
print(items[1])  # banana

```
