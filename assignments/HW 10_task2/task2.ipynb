Напишіть функцію, яка приймає два числа від користувача через input(), викликає числа a та b , а потім повертає значення квадрата a, поділеного на b , та створює блок try-except, який перехоплює виняток, якщо два значення, надані функцією input, не були числами, і якщо значення b дорівнювало нулю (не можна ділити на нуль). 

```python
def calculate():
    try:
        # Отримуємо числа від користувача
        a = float(input("Введіть число a: "))
        b = float(input("Введіть число b: "))

        # Обчислюємо квадрат a, поділений на b
        result = (a ** 2) / b
        print(f"Результат: {result}")
        return result

    except ValueError:
        # Якщо введено не число
        print("Помилка: потрібно вводити лише числа.")
    except ZeroDivisionError:
        # Якщо b = 0
        print("Помилка: не можна ділити на нуль.")

# Виклик функції
calculate()
```
