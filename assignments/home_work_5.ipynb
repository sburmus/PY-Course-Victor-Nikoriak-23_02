
# Task 1 — String manipulation
 ==============================

```python
user_string = input("Enter a string: ")

if len(user_string) < 2:
    result = ""
else:
    result = user_string[:2] + user_string[-2:]

print("Result:", result)
```

# Task 2 — Phone number validator
 ==============================

```python
phone = "1234567890"

if phone.isdigit() and len(phone) == 10:
    print("Valid phone number")
else:
    print("Invalid phone number")
```

```python
phone = "           1234567890"
print(len(phone))
phone_proces = phone.strip()
print(len(phone_proces))
if phone_proces.isdigit() and len(phone_proces) == 10:
    print("Valid phone number")
else:
    print("Invalid phone number")
```
